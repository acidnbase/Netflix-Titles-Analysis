## 1.What are the most popular genres on Netflix, and how has their popularity evolved over time?

```python
#Extract genres from the "listed_in" column
df['listed_in'] = df['listed_in'].apply(lambda x: x.split(','))
df_genres = df.explode('listed_in')

# Group data by year and genre and calculate the count of titles in each genre for each year
genre_counts_by_year = df_genres.groupby(['release_year', 'listed_in']).size().unstack(fill_value=0)

# Plot the popularity of genres over time
plt.figure(figsize=(12, 8))
genre_counts_by_year.plot(kind='line', marker='o', linewidth=2)
plt.title('Popularity of Genres on Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 3.Which countries produce the most content available on Netflix? Are there noticeable differences in the type of content (movies vs. TV shows)?

```python
ax = df.country.value_counts().nlargest(5).sort_values(ascending=True).plot(kind='barh',  color='lightgray')

bars = ax.patches
bars[-2].set_facecolor('#004c99')
```

```python
index = []
row = []
for c in ['United States', 'India', 'United Kingdom', 'Japan', 'South Korea'][::-1]:
  cnt = df[df.country==c].type.value_counts()  
  total = cnt['Movie'] + cnt['TV Show']
  index.append(c)
  row.append([cnt['Movie']/total*100, cnt['TV Show']/total*100,] )
cnt = pd.DataFrame(row, columns=['Movie', 'TV Show'], index=index)
cnt.plot(kind='barh', stacked=True)
```

## 5. What are the common durations for movies and TV shows? Is there a trend in the duration of content over the years?

```python
movie = df[df.type=='Movie'].copy()
movie.duration = movie.duration.str.split(expand=True)[0]
movie.duration = pd.to_numeric(movie.duration)
movie.duration.plot(kind='hist')
```

```python
def len_cat(x):
    if x <= 60:
        return "1"
    elif x <= 90:
        return "1.5"
    elif x <= 120:
        return "2"
    elif x <= 180:
        return "3"
    else:
        return ">3"
        
movie['len_cat'] = movie.duration.apply(len_cat)
movie['len_cat'].value_counts().plot(kind='bar')
```

## 7.Who are the most prolific directors and actors/actresses on Netflix? Are there any notable patterns or collaborations?

```python
# Split the "director" and "cast" columns to extract individual names
df['director'] = df['director'].fillna('').apply(lambda x: x.split(','))
df['cast'] = df['cast'].fillna('').apply(lambda x: x.split(','))

# Flatten the list of directors, actors, and actresses
directors = [director.strip() for sublist in df['director'] for director in sublist if director.strip()]
actors = [actor.strip() for sublist in df['cast'] for actor in sublist if actor.strip()]

# Count the occurrences of each director, actor, and actress
director_counts = pd.Series(directors).value_counts().head(10)
actor_counts = pd.Series(actors).value_counts().head(10)

# Plot the most prolific directors and actors/actresses
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
director_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Prolific Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
actor_counts.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Most Prolific Actors/Actresses on Netflix')
plt.xlabel('Actor/Actress')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 8.How has the rate of content addition to Netflix changed over time? Are there any seasonal patterns or spikes in new releases?

```python

# Convert the "date_added" column to datetime type and extract year and month
df['date_added'] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Group data by year and month and calculate the count of titles added each month
content_addition_by_month = df.groupby(['year_added', 'month_added']).size().reset_index(name='count')

# Plot the rate of content addition over time
plt.figure(figsize=(12, 6))
plt.plot(content_addition_by_month['year_added'], content_addition_by_month['count'], marker='o', color='skyblue')
plt.title('Rate of Content Addition to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.tight_layout()
plt.show()
```
